## HTTP/1.1 vs HTTP/2

```python
url = "https://es.investing.com/equities/bbva-historical-data"

head = {
    "User-Agent": 'Mozilla/5.0 (Windows NT 5.1; rv:31.0) Gecko/20100101 Firefox/31.0',
    "X-Requested-With": "XMLHttpRequest",
    "Accept": "text/html",
    "Accept-Encoding": "gzip, deflate, br",
    "Connection": "keep-alive",
}
```

### using requests HTTP/1.1

```python
import requests

req = requests.get(url, headers=head, timeout=5)
print(req.text)
```

### using hyper HTTP/2

```python
from hyper import HTTP20Connection

conn = HTTPConnection('www.investing.com:443')
conn.request('GET', '/equities/bbva-historical-data')
resp = conn.get_response()

print(resp.read())
```

```python
from hyper import HTTPConnection

conn = HTTPConnection('www.investing.com:443')
conn.request('GET', '/equities/bbva-historical-data')
resp = conn.get_response()

print(resp.read())
```

```python
from hyper import HTTP11Connection

req = HTTP11Connection('es.investing.com:443')
req.request('GET', '/equities/acciona-sa-historical-data')
resp = req.get_response()

print(resp.read())
```
